```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
plt.scatter(x, y)
m=x.shape[0]
```

```python
def cost(y_hat, y):
    l = (np.abs(y_hat-y))**3
    cost = np.mean(l)
    return cost
```

```python
def hypothesis(x, w, b):
    return np.multiply(x, w) + b
```

```python
def gradient_descent(x, y, w, b, learning_rate, m):
    y_hat = hypothesis(x, w, b)
    dw = (3.0/m) * np.sum(np.multiply(x,((y_hat-y)**2)))
    db = (3.0/m) * np.sum((y_hat-y)**2)
    w = w -learning_rate * dw
    b = b - learning_rate * db
    return w,b
```

```python
# Random Initialisation
w = np.random.randn(1)*0.01
b = np.random.randn(1)*0.01

# Hyperparameters
learning_rate = 0.0000002
num_epochs = 100000

L=[]
for i in range(num_epochs):
    w, b = gradient_descent(x,y,w,b,learning_rate, m)
    y_hat = hypothesis(x,w,b)
    L.append(cost(y_hat,y))
    if (i%(num_epochs/10))==0:
        plt.plot(x,y,'bo')
        plt.plot(x,y_hat,'g--')
        plt.show()
```

```python
print(f"Obtained value of w: {w}")
print(f"Obtained value of b: {b}")
print(f"Obtained line: y={w}x+{b}")
plt.plot(x,y,'bo')
plt.plot(x,y_hat,'g--')
plt.show()
```
